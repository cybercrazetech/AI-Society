```python
#Setting up libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#Setting up single variable dataset (y=mx+c)
#df = pd.read_csv("/path/to/csv")
#df = pd.read_excel("/path/to/excel")
data = {'area':  [2600,3000,3200,3600,4000],
        'price': [550000,565000,610000,680000,725000]}

df = pd.DataFrame(data)

df
```

```python
#Plotting graph
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
#building model
model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)
```

```python
#y=mx+c, showing results for m and c
print("m =",model.coef_,", c =",model.intercept_)
print("y =",str(model.coef_[0]),"x +",str(model.intercept_))
```

```python
#observe our best fit line
%matplotlib inline
plt.xlabel('area(sqr ft)')
plt.ylabel('price(US$)')
plt.scatter(df.area,df.price,color='red',marker='+')
plt.plot(df.area,model.predict(df[['area']]),color='blue')
```

```python
#prediction
model.predict([[3300]])
```

```python
#Summary(final code):

import pandas as pd
import numpy as np
from sklearn import linear_model

#df = pd.read_csv("/path/to/csv")
#df = pd.read_excel("/path/to/excel")
data = {'area':  [2600,3000,3200,3600,4000],
        'price': [550000,565000,610000,680000,725000]}

df = pd.DataFrame(data)

model = linear_model.LinearRegression()
model.fit(df[['area']],df.price)

model.predict([[3300]])
```
